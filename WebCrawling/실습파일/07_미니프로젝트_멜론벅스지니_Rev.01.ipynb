```python
from selenium import webdriver
```

```python
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
```

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)
```

```python
driver.get('https://www.melon.com/chart/index.htm')
```

```python
import threading
import pandas as pd

df = pd.DataFrame(columns=["title","artist"])
```

```python
# 멜론 1~50위 타이틀 리스트

titles = driver.find_elements(By.CSS_SELECTOR, '#lst50 > td:nth-child(6) > div > div')

title_result = []
for title in titles:
    title_list = title.find_element(By.CSS_SELECTOR,'div.ellipsis.rank01 > span > a').text
    title_result.append(title_list)
```

```python
# 멜론 1~50위 아티스트 리스트

artists = driver.find_elements(By.CSS_SELECTOR, '#lst50 > td:nth-child(6) > div > div')

artist_result = []

for artist in artists:
    artist_list = artist.find_element(By.CSS_SELECTOR, 'div.ellipsis.rank02 > a').text
    artist_result.append(artist_list)
```

```python
# 멜론 51~100위 타이틀 리스트

titles_100 = driver.find_elements(By.CSS_SELECTOR, '#lst100 > td:nth-child(6) > div > div')

title_100_result = []

for title_100 in titles_100:
    title_100_list = title_100.find_element(By.CSS_SELECTOR, 'div.ellipsis.rank01 > span > a').text
    title_100_result.append(title_100_list)
```

```python
# 멜론 51~100위 아티스트 리스트

artists_100 = driver.find_elements(By.CSS_SELECTOR, '#lst100 > td:nth-child(6) > div > div')

artist_100_result = []

for artist_100 in artists_100:
    artist_100_list = artist_100.find_element(By.CSS_SELECTOR, 'div.ellipsis.rank02 > a').text
    artist_100_result.append(artist_100_list)
```

```python
# 멜론 1~100위 타이틀, 아티스트 리스트 합치기

title_result.extend(title_100_result)
artist_result.extend(artist_100_result)
```

```python
# 멜론 데이터프레임 완성

df.loc[:,'title']=title_result
df.loc[:,'artist']=artist_result

df.insert(0, "rank", range(1,101))
df.insert(0, "platform", "melon")

df_melon = df
df_melon
```

```python
driver.get('https://music.bugs.co.kr/chart')
```

```python
# 벅스 타이틀 리스트
# #CHARTrealtime > table > tbody > tr:nth-child(n) > th > p > a

for n in range(1,100):
    titles_bugs = driver.find_elements(By.CSS_SELECTOR, '#CHARTrealtime > table > tbody > tr:nth-child(n)')

title_bugs_result = []
for title_bugs in titles_bugs:
    title_bugs_list = title_bugs.find_element(By.CSS_SELECTOR, 'th > p > a').text
    title_bugs_result.append(title_bugs_list)
```

```python
# 벅스 아티스트 리스트
# #CHARTrealtime > table > tbody > tr:nth-child(99) > td:nth-child(8) > p > a
# #CHARTrealtime > table > tbody > tr:nth-child(98) > td:nth-child(8) > p > a.artistTitle
# #CHARTrealtime > table > tbody > tr:nth-child(100) > td:nth-child(8) > p > a
# #CHARTrealtime > table > tbody > tr:nth-child(94) > td:nth-child(8) > p > a

for n in range(1,101):
    arists_bugs = driver.find_elements(By.CSS_SELECTOR, '#CHARTrealtime > table > tbody > tr:nth-child(n)')

artist_bugs_result = []
for artist_bugs in arists_bugs:
    artist_bugs_list = artist_bugs.find_element(By.CSS_SELECTOR, 'td:nth-child(8) > p > a').text
    artist_bugs_result.append(artist_bugs_list)
```

```python
# 벅스 데이터프레임 완성
df_bugs = pd.DataFrame({'title':title_bugs_result, 'artist':artist_bugs_result})
df_bugs.insert(0, 'rank', range(1,101))
df_bugs.insert(0, 'platform', 'bugs')
df_bugs
```

```python
# 지니 타이틀 리스트 (1~100위)
title_genie_result = []
artist_genie_result = []

def get_chart(page):
    driver.get("https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230702&hh=17&rtm=Y&pg=" + str(page))
    
    for n in range(1,101):
        titles_genie = driver.find_elements(By.CSS_SELECTOR, '#body-content > div.newest-list > div > table > tbody > tr:nth-child(n)')
    
    for title_genie in titles_genie:
        title_genie_list = title_genie.find_element(By.CSS_SELECTOR, 'td.info > a.title.ellipsis').text
        title_genie_result.append(title_genie_list)
        
        
    for n in range(1,101):
        artists_genie = driver.find_elements(By.CSS_SELECTOR, '#body-content > div.newest-list > div > table > tbody > tr:nth-child(n)')
    
    for artist_genie in artists_genie:
        artist_genie_list = artist_genie.find_element(By.CSS_SELECTOR, 'td.info > a.artist.ellipsis').text
        artist_genie_result.append(artist_genie_list)
        artist_genie_result
    
```
