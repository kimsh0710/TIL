```python
from selenium import webdriver
```

```python
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import threading
import pandas as pd
```

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)
```

```python
driver.get('https://www.melon.com/chart/index.htm')
```

```python
songs = driver.find_elements(By.CSS_SELECTOR, 'table > tbody > tr')

song_data = []
rank = 0
for song in songs:
    title_list = song.find_element(By.CSS_SELECTOR,'div.rank01 > span > a').text
    artist_list = song.find_element(By.CSS_SELECTOR,'div.rank02 > a').text
    rank = rank+1
    song_data.append(["melon",rank, title_list, artist_list])
    
    
df_melon = pd.DataFrame(song_data, columns=['서비스', '순위', '타이틀', '가수'])
df_melon
```

```python
driver.get('https://music.bugs.co.kr/chart')
```

```python
songs2 = driver.find_elements(By.CSS_SELECTOR, 'table.byChart > tbody > tr')
len(songs2)
```

```python
# 벅스 타이틀 리스트
# #CHARTrealtime > table > tbody > tr:nth-child(n) > th > p > a

songs2 = driver.find_elements(By.CSS_SELECTOR, 'table.byChart > tbody > tr')

song_data2 = []
rank = 0
for song2 in songs2:
    title_list2 = song2.find_element(By.CSS_SELECTOR,'p.title > a').text
    artist_list2 = song2.find_element(By.CSS_SELECTOR,'p.artist > a').text
    rank = rank+1
    song_data2.append(["bugs",rank, title_list2, artist_list2])
    
    
df_bugs = pd.DataFrame(song_data2, columns=['서비스', '순위', '타이틀', '가수'])
df_bugs
```

```python
driver.get('https://www.genie.co.kr/chart/top200')

songs3 = driver.find_elements(By.CSS_SELECTOR, 'table > tbody > tr')
len(songs3)
```

```python
# 지니 타이틀 리스트 (1~100위)

song_data3 = []


def get_chart(page):
    driver.get("https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230702&hh=17&rtm=Y&pg=" + str(page))
    
    songs3 = driver.find_elements(By.CSS_SELECTOR, 'table > tbody > tr')
    
    for song in songs3:
        title_list3 = song.find_element(By.CSS_SELECTOR, 'td.info > a.title').text
        artist_list3 = song.find_element(By.CSS_SELECTOR, 'td.info > a.artist').text
        song_data3.append(["genie",title_list3, artist_list3])
    
    
```

```python
%%time
for page in range(1,3):
    get_chart(page)
```

```python
df_genie = pd.DataFrame(song_data3, columns=['서비스', '타이틀', '가수'])
df_genie.insert(1, '순위', range(1,101))
df_genie
```

```python
new_df = pd.concat([df_melon,df_bugs],ignore_index=True)
new_df = pd.concat([new_df,df_genie],ignore_index=True)
new_df
```
