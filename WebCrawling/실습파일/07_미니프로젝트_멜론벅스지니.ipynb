### 벅스 사이트 분석

벅스 사이트 url : https://music.bugs.co.kr/chart

순위 태그 : .list trackList byChart > tbody > td > .ranking (column 1~100까지 추가해도 됨)

가수 : .list trackList byChart > tbody > .left > .artist > a ///p.artist > a


### 지니 사이트 분석

가수 : tbody > tr > .list > .info > .artist ellipsis ///a.artist ellipsis

```python
from selenium import webdriver
```

```python
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
```

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)
```

```python
driver.get('https://www.melon.com/chart/index.htm')
```

```python
import threading
import pandas as pd

df = pd.DataFrame(columns=["title","artist"])
```

```python
#lst50 > td:nth-child(6) > div > div > div.ellipsis.rank01 > span > a
```

```python
#titles = driver.find_elements(By.CSS_SELECTOR, 'div.ellipsis.rank01')
titles = driver.find_elements(By.CSS_SELECTOR, '#lst50 > td:nth-child(6) > div > div')

titles
```

```python
titles[0].find_element(By.CSS_SELECTOR, 'div.ellipsis.rank01 > span > a').text
```

```python
title_result = []
for title in titles:
    title_list = title.find_element(By.CSS_SELECTOR,'div.ellipsis.rank01 > span > a').text
    title_result.append(title_list)
```

```python
title_result[:10]
```

```python
df.loc[:,'title']=title_result
```

```python
print(df)
```

```python
# #lst50 > td:nth-child(6) > div > div > div.ellipsis.rank02 > a
```

```python
artists = driver.find_elements(By.CSS_SELECTOR, '#lst50 > td:nth-child(6) > div > div')

artists
```

```python
artists[0].find_element(By.CSS_SELECTOR, 'div.ellipsis.rank02 > a').text
```

```python
artist_result = []

for artist in artists:
    artist_list = artist.find_element(By.CSS_SELECTOR, 'div.ellipsis.rank02 > a').text
    artist_result.append(artist_list)
```

```python
artist_result[:10]
```

```python
df.loc[:,'artist']=artist_result
print(df)
```

```python
df_melon = df
```

```python
df_melon
```

```python
# #lst100 > td:nth-child(6) > div > div > div.ellipsis.rank01 > span > a

titles_100 = driver.find_elements(By.CSS_SELECTOR, '#lst100 > td:nth-child(6) > div > div')
```

```python
title_100_result = []

for title_100 in titles_100:
    title_100_list = title_100.find_element(By.CSS_SELECTOR, 'div.ellipsis.rank01 > span > a').text
    title_100_result.append(title_100_list)
    
title_100_result[:10]
```

```python
# #lst100 > td:nth-child(6) > div > div > div.ellipsis.rank02 > a

artists_100 = driver.find_elements(By.CSS_SELECTOR, '#lst100 > td:nth-child(6) > div > div')
```

```python
artist_100_result = []

for artist_100 in artists_100:
    artist_100_list = artist_100.find_element(By.CSS_SELECTOR, 'div.ellipsis.rank02 > a').text
    artist_100_result.append(artist_100_list)
    
artist_100_result[:10]
```

```python
melon_100 = {'title' : title_100_result, 'artist' : artist_100_result}
df_melon_100 = pd.DataFrame(melon_100)
df_melon_100
```

```python
new_df_melon = pd.concat([df_melon,df_melon_100])
new_df_melon
```

```python
new_df_melon.insert(0,"rank",range(1,101))
new_df_melon
```
