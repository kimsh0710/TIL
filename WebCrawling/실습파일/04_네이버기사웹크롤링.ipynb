```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://news.naver.com/"
html = urlopen(url)
```

```python
news_html = BeautifulSoup(html, "html.parser")
```

```python
print(news_html.prettify())
```

```python
name = news_html.find('ul', {"class":"Nlnb_menu_list"})
name
```

```python
lis = name.find_all('li')
lis
```

```python
len(lis)
```

```python
for li in lis:
    a_tag = li.find('a')
    print(a_tag.text + ":" + a_tag["href"])
    
```

```python
for li in lis:
    a_tag = li.find('a')
    print(a_tag.text)
    print(a_tag["href"])
```

```python
section = []
link = []
```

```python
for li in lis:
    a_tag = li.find('a')
    section.append(a_tag.text)
    link.append(a_tag["href"])
    
```

```python
import pandas as pd
col_dict = {'section':section, "link":link}
news_section_df = pd.DataFrame(col_dict)
news_section_df
```

```python
news_section_df.to_csv(r'C:\JupyterTest\naver_news_section.csv', encoding='euc-kr')
```
